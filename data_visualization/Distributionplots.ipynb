We use distribution plots to visualize the distribution of quantitative data

Let's discuss some plots that allow us to visualize the distribution of a dataset. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'], bins=20, kde=True)
```

```python
sns.histplot(df['total_bill'], bins=20, kde=True)
```

```python
plt.figure(figsize=(12, 4))  # (width, height) in inches

plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins=20, kde=True)

plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins=20, kde=True)
```

```python
# for comparison we use joint plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='resid')
```

```python
sns.pairplot(df, hue='size', palette='rainbow') # comparison of all numeric colummns
```

```python
sns.rugplot(df['total_bill'])
```
