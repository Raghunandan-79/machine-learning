### 1. **Matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### **Matplotlib Basics**

```python
x = np.linspace(0, 5, 11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

#### **Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x, y**2)

plt.subplot(2,2,4)
plt.plot(x, y)

plt.subplot(2,2,3)
plt.plot(y, x)

plt.subplot(2,2,2)
plt.plot(x**2, x**2)
```

#### **Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the figure object is created. You can use the figsize and dpi keyword arguments.

Figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inc (pixel per inch)

```python
fig = plt.figure(figsize=(12,8), dpi=200) # 1200 x 800

# left, bottom, width, heighta
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.2])
axes2.plot(x, y)
```

#### **Saving Plots**

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("output/myplot.png", dpi=200, bbox_inches='tight')
```
