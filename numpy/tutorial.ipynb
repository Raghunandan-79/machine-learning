### Origin story of numpy

python was always very powerful with maths but one thing was missing that was arrays to 
solve this thing there was a creater named as James Hugunin in 1995 and he created a numeric
library. This library provided array support but the problem was the performance was not 
that good and thing thing was further resolved in early 2000's another library was developed
by Space Telescope Science Intitute and named it numarray to increase the power of previously
created library but it was not compatible with numeric library

After that in 2005 there was a professor and scientist named as Travis Oliphant who merged
numeric and numarray and named it Numpy and finally there was library named as numpy

It was made in C thats why it is 50x faster than python list

### Numpy features
    - Data Type: Same data type only
    - Performance: Much faster (c-level)
    - Memory efficency: High
    - Vector operations: Easy + fast

### Numpy Indexing and Slicing of vectors

```python
arr = np.arange(11, 21)
arr
```

```python
print(arr[6])
```

```python
print(arr[1:5])
```

```python
print(arr[:5])
```

```python
print(arr[:])
```

```python
print(arr[3:])
```

```python
print(arr[3::2])
```

### Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1, 31).reshape(6, 5)
arr
```

```python
print(arr[5]) # it will target rows
```

```python
print(arr[5, 3]) # first number will target rows and second number will target columns
```

```python
print(arr[0:2, 1:3])
```

```python
print(arr[3:, 3:])
```

```python
print(arr[:, 2]) # indexing a column
```
