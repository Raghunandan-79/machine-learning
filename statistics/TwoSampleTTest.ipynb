```python
import numpy as np
from scipy import stats
```

```python
group_a = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88]
group_b = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]
```

```python
t_stats, p_value = stats.ttest_ind(a=group_a, b=group_b, equal_var=False)
```

```python
print(t_stats)
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null test")
```
