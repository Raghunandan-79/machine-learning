**Q1: What is Regression ?**

**Ans** Regression is a way to predict a number. It helps us find out how one thing changes when something else changes.

**Steps**

1. Start
2. Collect data
3. Define variables
4. Split Dataset
5. Train Model
6. Calculate Coefficients
7. Make Predictions
8. Evaluate model
9. Model Satisfactory
10. If yes deploy model else adjust model or data and go back to Step 4 an repeat process

**Ans** Suppose we have variable age and salary. Now what the question arises is can you predict the salary of the person according to age. Here age is input variable and salary is output variable

Imagine you are running a lemonade stand. You've been keeping track of:
- How many glasses you sell each day
- And what the temperature was on that day

After a while, you notice:

**"The hotter the day, the more lemonade I sell."**

Regression is the math that helps you figure out exactly how much more lemonade you'll sell as the temperature goes up.

**Regression is about finding patterns in numbers so we can make smart predictions**
